```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
sales_df = pd.read_csv("sales_data.csv")
churn_df = pd.read_csv("customer_churn.csv")

# Display first 5 rows
print("Sales Data:")
print(sales_df.head())

print("\nCustomer Churn Data:")
print(churn_df.head())
```

```python
print("Sales Data:")
print(sales_df.head())

print("\nSales Data Info:")
print(sales_df.info())

print("\nCustomer Churn Data:")
print(churn_df.head())

print("\nCustomer Churn Info:")
print(churn_df.info())
```

```python
# Check missing values
print("Missing Values in Sales Data:")
print(sales_df.isnull().sum())

print("\nMissing Values in Customer Churn Data:")
print(churn_df.isnull().sum())

# Remove duplicate rows
sales_df = sales_df.drop_duplicates()
churn_df = churn_df.drop_duplicates()

print("\nSales Data Shape:", sales_df.shape)
print("Customer Churn Data Shape:", churn_df.shape)
```

```python
print("Sales Data Columns:")
for col in sales_df.columns:
    print(col)

print("\nCustomer Churn Columns:")
for col in churn_df.columns:
    print(col)
```

```python
# SALES ANALYSIS

# Total Sales
print("Total Sales:", sales_df["Total_Sales"].sum())

# Average Sales
print("Average Sales:", sales_df["Total_Sales"].mean())

# Top Products
top_products = sales_df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)
print("\nTop Products:")
print(top_products.head())

# Sales by Region
region_sales = sales_df.groupby("Region")["Total_Sales"].sum().sort_values(ascending=False)
print("\nSales by Region:")
print(region_sales)

# Top Customers
top_customers = sales_df.groupby("Customer_ID")["Total_Sales"].sum().sort_values(ascending=False)
print("\nTop Customers:")
print(top_customers.head())
```

```python
# VISUALIZATION 1

top_products.head().plot(kind="bar")
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
# VISUALIZATION 2

region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```
